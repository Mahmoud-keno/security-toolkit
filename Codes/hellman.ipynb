```python
# Step 1: Generate public parameters (prime p and base g)
# For simplicity, we use small numbers from the example
p = 23
g = 5
print("Public Parameters:")
print(f"Prime p = {p}")
print(f"Base g = {g}")
```

```python
# Step 2: Input private keys for Alice and Bob
# Alice's private key
a = int(input("Enter Alice's private key (a): "))
# Bob's private key  
b = int(input("Enter Bob's private key (b): "))

print(f"\nAlice's private key: {a}")
print(f"Bob's private key: {b}")
```

```python
# Step 3: Compute public keys
# Alice computes A = g^a mod p
A = pow(g, a, p)
# Bob computes B = g^b mod p
B = pow(g, b, p)

print(f"\nPublic Keys:")
print(f"Alice's public key A = g^a mod p = {g}^{a} mod {p} = {A}")
print(f"Bob's public key B = g^b mod p = {g}^{b} mod {p} = {B}")
```

```python
# Step 4: Exchange public keys (simulated)
print("\nExchanging public keys...")
```

```python
# Step 5: Compute shared secret
# Alice computes s = B^a mod p
s_alice = pow(B, a, p)
# Bob computes s = A^b mod p
s_bob = pow(A, b, p)

print(f"\nShared Secret Calculation:")
print(f"Alice computes: s = B^a mod p = {B}^{a} mod {p} = {s_alice}")
print(f"Bob computes: s = A^b mod p = {A}^{b} mod {p} = {s_bob}")
```

```python
# Step 6: Verify both secrets match
if s_alice == s_bob:
    print(f"\n✓ Shared secret established: {s_alice}")
    print("Alice and Bob can now use this for symmetric encryption.")
else:
    print("\n✗ Error: Shared secrets do not match!")
```
